```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
%matplotlib inline
```

```python
data= pd.read_csv('Avalanche.csv', delimiter=',')
```

```python
data
```

```python
data.isnull().any()
```

```python
y= data['Prediction'].values
y
```

```python
x= data.drop('Prediction', axis=1)
```

```python
x.head()
```

```python
X= x.values
X
```

```python
ct= ColumnTransformer([("oh",OneHotEncoder(),[1])],remainder= "passthrough")
```

```python
X=ct.fit_transform(X)
```

```python
X= X[:,1:]
X
```

```python
X1=X[:,0:2]
X1
```

```python
#normalising the data
sc= StandardScaler()
X= sc.fit_transform(X[:,2:6])
X
```

```python
X= np.hstack((X1,X))
X
```

```python
##spliting into train and test data
x_train, x_test, y_train, y_test= train_test_split(X,y, train_size=0.8, random_state=0)
```

### Building the model

```python
#decision tree
avalanche_dt= DecisionTreeClassifier(criterion= 'entropy', random_state=0)
avalanche_dt.fit(x_train, y_train)
```

```python
y_pred_dt= avalanche_dt.predict(x_test)
```

```python
accuracy_score(y_test, y_pred_dt)
```

```python
confusion_matrix(y_test, y_pred_dt)
```

```python
y_val= avalanche_dt.predict(x_train)
accuracy_score(y_train, y_val)
```

```python
#random forest
rf= RandomForestClassifier(n_estimators=10, criterion= 'entropy', random_state=0)
rf.fit(x_train,y_train)
```

```python
y_pred_rf= rf.predict(x_test)
```

```python
accuracy_score(y_test, y_pred_rf)
```

```python
confusion_matrix(y_test, y_pred_rf)
```

```python
y_val2= rf.predict(x_train)
accuracy_score(y_train, y_val2)
```
